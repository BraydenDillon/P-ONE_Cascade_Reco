```python
gcdfile = dataio.I3File('../gcdfile/PONE_800mGrid.i3.gz')
```

```python
for frame in gcdfile:
    gcdframe = frame
    break
```

```python
strings = []
string_x = []
string_y = []
for dom in gcdframe['I3ModuleGeoMap']:
    if dom.key().string in strings:
        continue
    strings.append(dom.key().string)
    string_x.append(gcdframe['I3ModuleGeoMap'][dom.key()].pos.x)
    string_y.append(gcdframe['I3ModuleGeoMap'][dom.key()].pos.y)


randstr_idx = np.random.randint(0, len(strings), 80)
rand_x = []
rand_y = []
for i in randstr_idx:
    rand_x.append(string_x[i])
    rand_y.append(string_y[i])

plt.scatter(string_x, string_y, alpha=0.3)
plt.scatter(rand_x, rand_y, color = 'red')

    
```
